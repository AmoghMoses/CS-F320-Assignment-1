```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
print("Success")
```

```python
a,b = 2, 2
print("(a,b) = ","(",a,",", b,')', sep = '')
x = np.linspace(beta.ppf(0.0, a, b),beta.ppf(1.0, a, b),100)     # Linearly spacing points between x and y and num = 100
plt.figure(figsize=(7,7))                                        # Sets the height and width of the graph with parameters in the form of (width,height)
plt.plot(x, beta.pdf(x,a,b), 'b-')                               # plot takes two variables x,y. Here y is the beta function with two parameters a,b and r- is for the color
plt.title('Beta Distribution', fontsize='15')                    # Gives teh title for the graph
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='15')  # Gives the title for the x-axis
plt.ylabel('Probability', fontsize='15')                         # Gives the title for the y-axis
plt.show() 
```

```python
a+=40                                                            # Changing the value of 'a' based on question
b+=10                                                            # Changing the value of 'b' based on question
print("(a,b) = ","(",a,",", b,')', sep = '')
plt.figure(figsize=(7,7))
plt.plot(x, beta.pdf(x, a, b), 'b-')
plt.title('Beta Distribution', fontsize='15')
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='15')
plt.ylabel('f(x)', fontsize='15')
plt.show()
```
